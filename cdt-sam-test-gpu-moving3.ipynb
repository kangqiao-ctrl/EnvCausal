```python
df_c1 = pd.read_csv('../../latest_dataset/ts/3day/df_m3_c1_metro_with_day.csv')
```

```python
df_c1=df_c1.drop(columns=['Unnamed: 0','date','city','MORB%','log(Case+1)','log(MORB%+1)'])
```

```python
from cdt.causality.graph import SAM
```

```python
model1 = SAM(njobs=2, gpus=1,nruns=4)
```

```python
model1= SAM()
```

```python
torch.multiprocessing.set_start_method('spawn', force=True)
```

```python
output = model1.predict(df_c1)
```

```python
output_c1 = output
```

```python
a = nx.draw_networkx(output_c1, font_size=12)
plt.show()
```

```python
am_c1 = nx.adjacency_matrix(output_c1).todense()
```
